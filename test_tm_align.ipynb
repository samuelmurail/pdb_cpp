```python
import pdb_numpy
from pdb_numpy import DSSP, alignement
#sys.path.insert(0, "./src")

from pdb_cpp import Coor, core
from pdb_cpp import TMalign
```

```python
file_name = "3eam.pdb"
file_name_2 = "5bkg.pdb"
```

```python
coor_1 = pdb_numpy.Coor(file_name)
coor_2 = pdb_numpy.Coor(file_name_2)
```

```python
rmsd, _ = alignement.align_seq_based(
        coor_1,
        coor_2,
        #chain_1=align_chains,
        #chain_2=align_chains,
    )
print(rmsd)
coor_1.write("aligned_3eam_numpy.pdb")
coor_2.write("aligned_5bkg_numpy.pdb")
```
